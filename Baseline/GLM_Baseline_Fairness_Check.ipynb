#### 공정성 검정

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import pickle

# 1. 모델 불러오기
file_name = 'churn_model_glm_final.pkl'
with open(file_name, 'rb') as f:
    loaded_pack = pickle.load(f)
    
base_model = loaded_pack['model']
best_threshold = loaded_pack['threshold']

# 폰트 설정
plt.rc('font', family='AppleGothic')
plt.rcParams['axes.unicode_minus'] = False

# 2. 예측 및 기본 분석 데이터프레임 구축
test_preds_proba = base_model.predict_proba(X_test)[:, 1]
test_preds_final = (test_preds_proba >= best_threshold).astype(int)

analysis_df = X_test.copy()
analysis_df['true_label'] = y_test.values
analysis_df['pred_label'] = test_preds_final

# 3. 민감 변수 구간 설정
sens_config = {
    'age': {
        'bins'  : [0, 20, 30, 40, 50, 60, 70, 120],
        'labels': ['10s', '20s', '30s', '40s', '50s', '60s', '70s+']
    },
    'age_gender': {
        'bins'  : [0, 26, 41, 61, 120],
        'labels': ['청년', '중년', '장년', '고령']
    },
    'C_GI': {
        'bins'  : [-2, 0, 25, 50, 75, 101],
        'labels': ['Unknown', 'GI_1', 'GI_2', 'GI_3', 'GI_4']
    },
    'C_IE_T': {
        'bins'  : [-2, 0, 25, 50, 75, 101],
        'labels': ['Unknown', 'C_IE_T_1', 'C_IE_T_2', 'C_IE_T_3', 'C_IE_T_4']
    }
}

# 각 변수별 파싱 및 문자열 변환
analysis_df['age_binned'] = pd.cut(analysis_df['age'], bins=sens_config['age']['bins'], labels=sens_config['age']['labels'], right=False).astype(str)
analysis_df['gender_str'] = analysis_df['gender'].astype(str)

# age_gender 결합
age_grp_temp = pd.cut(analysis_df['age'], bins=sens_config['age_gender']['bins'], labels=sens_config['age_gender']['labels'], right=False).astype(str)
analysis_df['age_gender_combined'] = age_grp_temp + ' ' + analysis_df['gender_str']

analysis_df['C_GI_binned'] = pd.cut(analysis_df['C_GI'], bins=sens_config['C_GI']['bins'], labels=sens_config['C_GI']['labels'], right=False).astype(str)
analysis_df['C_IE_T_binned'] = pd.cut(analysis_df['C_IE_T'], bins=sens_config['C_IE_T']['bins'], labels=sens_config['C_IE_T']['labels'], right=False).astype(str)

# [수정 포인트 1] KeyError 해결 및 역매핑 사전 적용
# 실제 피처 컬럼명인 'C_H_num'을 호출하고 원래 정의된 레이블(H1~H6)로 변환합니다.
c_h_map_inv = {500: 'H1', 5500: 'H2', 17500: 'H3', 37500: 'H4', 75000: 'H5', 150000: 'H6', -1: 'Unknown'}
analysis_df['C_H_str'] = analysis_df['C_H_num'].fillna(-1).astype(float).astype(int).map(c_h_map_inv).fillna('Unknown')

# 4. 분석 대상 6개 피처와 실제 컬럼 매핑 및 정렬 순서 정의
target_features = ['age', 'gender', 'age_gender', 'C_GI', 'C_IE_T', 'C_H']
feature_col_map = {
    'age': 'age_binned',
    'gender': 'gender_str',
    'age_gender': 'age_gender_combined',
    'C_GI': 'C_GI_binned',
    'C_IE_T': 'C_IE_T_binned',
    'C_H': 'C_H_str'
}

# FPR 계산 함수
def calculate_fpr(group):
    negatives = group[group['true_label'] == 0]
    if len(negatives) == 0:
        return np.nan
    return negatives['pred_label'].mean()

# 전체 평균 FPR 계산
total_fpr = analysis_df[analysis_df['true_label'] == 0]['pred_label'].mean() * 100

# 5. 6대 민감 변수 순회 분석 및 시각화 루프
for feature in target_features:
    col_name = feature_col_map[feature]
    
    # 가나다 순서 변형 차단용 정렬 순서 강제 주입
    if feature == 'age':
        custom_order = list(sens_config['age']['labels'])
    elif feature == 'gender':
        custom_order = ['F', 'M']
    elif feature == 'age_gender':
        custom_order = [f"{a} {g}" for a in sens_config['age_gender']['labels'] for g in ['F', 'M']]
    elif feature in ['C_GI', 'C_IE_T']:
        custom_order = list(sens_config[feature]['labels'])
    elif feature == 'C_H':
        # [수정 포인트 2] C_H 계열도 H1부터 H6까지 순서대로 정렬되도록 지정
        custom_order = ['H1', 'H2', 'H3', 'H4', 'H5', 'H6', 'Unknown']
        
    # 데이터셋에 실제로 존재하는 범주들만 순서에 남겨 필터링
    custom_order = [cat for cat in custom_order if cat in analysis_df[col_name].unique()]
    
    # 해당 컬럼을 명시적 Categorical 타입으로 변경하여 정렬 고정
    analysis_df[col_name] = pd.Categorical(analysis_df[col_name], categories=custom_order, ordered=True)
    
    # 그룹바이 통계량 요약 표 작성
    feature_fpr_df = analysis_df.groupby(col_name, observed=False).apply(
        lambda x: pd.Series({
            '전체_고객수'      : len(x),
            '유지_고객수(N)'   : len(x[x['true_label'] == 0]),
            '오탐지_고객수(FP)': len(x[(x['true_label'] == 0) & (x['pred_label'] == 1)]),
            'FPR(%)'          : calculate_fpr(x) * 100
        })
    ).reset_index()
    
    feature_fpr_df['FPR(%)'] = feature_fpr_df['FPR(%)'].round(2)
    
    # 표 출력
    print("\n" + "=" * 70)
    print(f"📊 [{feature}] 집단별 FPR(False Positive Rate) 비교 통계")
    print("=" * 70)
    print(feature_fpr_df.to_string(index=False, na_rep='-'))
    print(f"📢 전체 평균 FPR: {total_fpr:.2f}%")
    print("=" * 70)
    
    # 히스토그램 시각화
    plt.figure(figsize=(11, 5))
    bars = plt.bar(
        feature_fpr_df[col_name].astype(str),
        feature_fpr_df['FPR(%)'],
        color='lightcoral' if feature != 'age_gender' else 'teal',
        edgecolor='black',
        alpha=0.8,
        label=f'{feature} 집단별 FPR'
    )
    
    # 상단 수치(%) 데이터 레이블링
    for bar in bars:
        height = bar.get_height()
        if not np.isnan(height) and height > 0:
            plt.text(
                bar.get_x() + bar.get_width() / 2.,
                height + (total_fpr * 0.02),
                f'{height:.2f}%',
                ha='center', va='bottom', fontsize=9, fontweight='bold'
            )
            
    # 전체 평균 기준선 배치
    plt.axhline(
        y=total_fpr,
        color='royalblue',
        linestyle='--',
        linewidth=1.5,
        label=f'전체 평균 FPR: {total_fpr:.2f}%'
    )
    
    plt.title(f'[{feature}] 집단별 모델의 오탐지율(FPR) 비교 - 공정성 성적표', fontsize=14, fontweight='bold', pad=15)
    plt.xlabel(f'{feature} 범주', fontsize=11)
    plt.ylabel('FPR (%)', fontsize=11)
    plt.xticks(rotation=30, ha='right')
    plt.legend(loc='upper right')
    plt.grid(axis='y', linestyle=':', alpha=0.6)
    plt.tight_layout()
    plt.show()
```
